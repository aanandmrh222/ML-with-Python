```python
import numpy as np
from geneticalgorithm import geneticalgorithm as ga
```

```python
def f(X):
    penalty = 0
    if X[0]+X[1] < 2:
        penalty = 50+100*(2-X[0]-X[1])
    return np.sum(X)

algorithm_param = {'max_num_iteration':300, 'population_size':100, 'mutation_probability':0.1,
                  'elit_ratio':0.1, 'crossover_probability':0.5, 'parents_portion':0.3, 
                  'crossover_type':'uniform', 'max_iteration_without_improv':None}
varbound = np.array([[0,10]]*3)
model = ga(function=f, dimension=3, variable_type='real', variable_boundaries=varbound, 
           algorithm_parameters=algorithm_param)
model.run()
```
