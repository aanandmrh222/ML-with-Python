```python
import pandas as pd
```

```python
data = pd.read_csv('Data.csv')
data 
```

```python
# using .get_dummies function to convert 
# the categorical datatype to numerical and storing the returned dataFrame in a new variable
data1 = pd.get_dummies(data['Purchased'])

# using pd.concat to concatenate the dataframes 
# data and data1 and storing the concatenated
# dataFrame in data
data = pd.concat([data, data1], axis=1)

data.drop('Purchased', axis=1, inplace=True)
data
```

```python
data2 = pd.read_csv('Data.csv')
# importing LabelEncoder from SkLearn
# Library from preprocessing Module 
from sklearn.preprocessing import LabelEncoder

# Creating a instance of label Encoder
le = LabelEncoder()

# Using .fit_transform function to fit label
# encoder and return encoded label
label = le.fit_transform(data2['Purchased'])

# printing label 
label
```

```python
# data2.drop('Purchased', axis=1, inplace = True)
data2['Purchased'] = label
data2
```
